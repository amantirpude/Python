```python
import time
## insert card
print("Insert Your Card")

## delay time after inserting card
time.sleep(5)
password = 2903
# Choose language
language = print("""Please choose your language
    कृपया अपनी भाषा चुनें
    कृपया तुमची भाषा निवडा
    1 == English
    2 == Hindi
    """)
choice = int(input("Enter here "))

if choice == 1:
    pin = int(input("Please Enter Your 4 Digit Pin - "))

    balance = 10000
    if pin == password:
        while True:
            print(""" Please choose an option
            1 == Balance
            2 == Withdrawl amount
            3 == Deposite Amount
            4 == Exit
            """)
            try:
                option = int(input("Please Enter your Option"))
            except:
                print("Enter a valid Option")
            if option == 1:
                print("Your account balance is - ", balance)
                break
            elif option == 2:
                withdrawl_amt = int(input("Enter the amount to withdraw - "))
                print('Please collect your cash')
                time.sleep(5)
                print("{} is debited from your Account".format(withdrawl_amt))
                print("Account Balance is - ", balance-withdrawl_amt)
                break
            elif option == 3:
                deposit_amt = int(input("Please Enter your Deposit amount - "))
                print("Please wait...")
                time.sleep(5)
                print("{} is credited in your Account".format(deposit_amt))
                print("Account Balance is - ", balance + deposit_amt)
                break
            else:
                break
    else:
        print("Please Enter the correct pin")
elif choice == 2:
    pin = int(input("कृपया अपना 4 अंकों का पिन दर्ज करें"))

    balance = 10000
    if pin == password:
        while True:
            print(""" कृपया एक विकल्प चुनिए
            1 == बैलेंस
            2 == निकासी राशि
            3 == जमा राशि
            4 == बाहर निकलें
            """)
            try:
                option = int(input("कृपया अपना विकल्प दर्ज करें"))
            except:
                print("एक वैध विकल्प दर्ज करें")
            if option == 1:
                print("आपका खाता शेष है - ", balance)
                break
            elif option == 2:
                withdrawl_amt = int(input("निकालने के लिए राशि दर्ज करें - "))
                print('कृपया अपना कैश जमा करें ...')
                time.sleep(5)
                print("{} आपके खाते से डेबिट किया गया है".format(withdrawl_amt))
                print("खाता शेष है - ", balance-withdrawl_amt)
                break
            elif option == 3:
                deposit_amt = int(input("कृपया अपनी जमा राशि दर्ज करें"))
                print("कृपया प्रतीक्षा करें...")
                time.sleep(5)
                print("{} आपके खाते में जमा किया जाता है".format(deposit_amt))
                print("खाता शेष है - ", balance + deposit_amt)
                break
            else:
                break
    else:
        print("कृपया सही पिन दर्ज करें")
else:
    print("""Wrong choice
    Please choose your language
    कृपया अपनी भाषा चुनें
    कृपया तुमची भाषा निवडा
    1 == English
    2 == Hindi
    """)
```
